```python
import pandas as pd
from google.colab import files

uploaded = files.upload()

# Get the uploaded file name (assuming only one file is uploaded)
uploaded_file_name = list(uploaded.keys())[0]

# Read the Excel file
df = pd.read_excel(uploaded_file_name)

print("Original Data:")
print(df)

df.fillna(df.mean(numeric_only=True), inplace=True)

print("\nData After Filling Missing Values:")
print(df)

high_achievers = df[(df['Math'] > 75) & (df['Science'] > 75) & (df['English'] > 75)]

print("\nHigh Achievers (Score > 75 in all subjects):")
print(high_achievers)

# You might want to save these as Excel files too, or specify an encoding if saving as CSV
df.to_csv("cleaned_student_scores.csv", index=False, encoding='utf-8')
high_achievers.to_csv("high_achievers.csv", index=False, encoding='utf-8')


files.download("cleaned_student_scores.csv")
files.download("high_achievers.csv")
```

```python
import pandas as pd
from google.colab import files

# Upload file
uploaded = files.upload()

# Get the uploaded file name (assuming only one file is uploaded)
uploaded_file_name = list(uploaded.keys())[0]

# Read the Excel file
df = pd.read_excel(uploaded_file_name)

# Fill missing Age values with median
df['Age'] = df['Age'].fillna(df['Age'].median())

# Filter elderly female patients diagnosed with Diabetes
elderly_female_diabetes = df[(df['Gender'] == "Female") & (df['Age'] > 50) & (df['Diagnosis'].str.lower() == "diabetes")]

# Save results
df.to_csv("cleaned_patients.csv", index=False)
elderly_female_diabetes.to_csv("elderly_female_diabetes.csv", index=False)

# Download results
files.download("cleaned_patients.csv")
files.download("elderly_female_diabetes.csv")
```
